```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('girl.jpg')

lower_reso = cv2.pyrDown(img)

plt.imshow(img)
plt.title('Original')
plt.show()

plt.imshow(lower_reso)
plt.title('Image Pyramids')
plt.show()

cv2.imshow('Image Pyramids',lower_reso)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
