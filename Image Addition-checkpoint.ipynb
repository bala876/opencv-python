```python
from matplotlib import pyplot as plt
import cv2
import numpy as np

img1 = cv2.imread('1.jpg')
img2 = cv2.imread('2.jpg')
print(img1.shape)
print(img2.shape)
dst = cv2.addWeighted(img1,0.7,img2,0.3,0)

plt.imshow(dst)
plt.title('Adding images')
plt.show()
```
