```python
# 1. Displaying an Image

from matplotlib import pyplot as plt
import cv2

img = cv2.imread('new.jpg', cv2.IMREAD_COLOR)

plt.imshow(img)
plt.title('My Picture')
plt.show()
```

```python
# 2. Cropping a part of the image using slicing numpy arrays

print(img.shape)
ball = img[120:200,200:300]
plt.imshow(ball)
plt.title('My Picture')
plt.show()
```

```python
ball = img[120:200,200:300]
img[20:100,100:200] = ball
plt.imshow(img)
plt.title('My Picture')
plt.show()
```
