```python
# 1. Constant Border
```

```python
from matplotlib import pyplot as plt
import cv2
x = [255,0,0]
img = cv2.imread('girl.jpg', cv2.IMREAD_COLOR)

replicate = cv2.copyMakeBorder(img,50,50,50,50,cv2.BORDER_CONSTANT,value=x)

plt.imshow(replicate)
plt.title('My Picture')
plt.show()
```

```python
from matplotlib import pyplot as plt
import cv2
x = [255,0,0]
img = cv2.imread('girl.jpg', cv2.IMREAD_COLOR)

replicate = cv2.copyMakeBorder(img,50,50,50,50,cv2.BORDER_REFLECT_101)

plt.imshow(replicate)
plt.title('My Picture')
plt.show()
```
