```python
Now check the different shapes of the image
```

```python
import cv2
import numpy as np
 
img1 = cv2.imread('new.jpg', cv2.IMREAD_COLOR)            # BGR
img2 = cv2.imread('new.jpg', cv2.IMREAD_UNCHANGED)        # RGBA
img3 = cv2.imread('new.jpg', cv2.IMREAD_GRAYSCALE)        # GrayScale

print(img1.shape)
print(img2.shape)
print(img3.shape)
```

```python
#Changing the Multiple Values of the pixel
```

```python
from matplotlib import pyplot as plt
import cv2

img = cv2.imread('new.jpg', cv2.IMREAD_COLOR)

plt.imshow(img)
plt.title('My Picture')
plt.show()
```

```python
# With Changing

from matplotlib import pyplot as plt
import cv2

img = cv2.imread('new.jpg', cv2.IMREAD_COLOR)

for i in range(0,200):
    for j in range(0,200):
        img[i][j] = [255, 255, 255]

plt.imshow(img)
plt.title('My Picture')
plt.show()
```
