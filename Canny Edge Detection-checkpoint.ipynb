```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('girl.jpg')

edges = cv2.Canny(img,100,200)

plt.imshow(img)
plt.title('Original')
plt.show()

plt.imshow(edges)
plt.title('Canny Edge Detection')
plt.show()

cv2.imshow('Canny Edge Detection',edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
